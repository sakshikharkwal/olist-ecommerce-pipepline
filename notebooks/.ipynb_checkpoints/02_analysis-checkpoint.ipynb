### ORDERS

#### AVERAGE DELIVERY TIME

```python
avg_del_time = orders_i['delivery_time'].mean()
print(f"Average delivery time: {avg_del_time:.2f} hours")
```

#### PROPORTION OF LATE DELIVERIES

```python
late_deliveries = (orders_i['delivery_delay']>0).mean()
print(f"Proportion of late deliveries: {late_deliveries:.2%}")
```

#### DELIVERY DELAY DISTRIBUTION

```python
plt.figure(figsize=(8,4))
orders['delivery_delay'].hist(bins=30, color='hotpink')
plt.title("Delivery Delay Distribution")
plt.xlabel("Delay (hours)")
plt.ylabel("Number of orders")
plt.show()
```

#### ORDER STATUS ANALYSIS

```python
plt.figure(figsize=(8,6))
orders_i['order_status'].value_counts().plot(kind='bar', color='skyblue')
plt.xlabel('Order Status')
plt.ylabel('Number of Orders')
plt.title('Order Status Analysis')
plt.xticks(rotation=45)
plt.show()
```

#### CANCELLATION RATE

```python
cancel_rate = (orders_i["order_status"].eq("canceled").sum() / len(orders_i)) * 100
print(f"Cancellation Rate: {cancel_rate:.2f}%")
```

<div class = 'alert alert-block alert-success'>
    Very low cancellation rate indicates a reliable platform.
</div>

#### DELIVERY PERFORMANCE

```python
datetime_features_orders = [feature for feature in orders_i.columns if 'order' in feature and feature not in ['order_id', 'order_status']]
datetime_features_orders
```

```python
for feature in datetime_features_orders:
    orders_i[feature] = pd.to_datetime(orders_i[feature], errors='coerce')
orders_i.dtypes
```

```python
datetime_features_reviews = ['review_creation_date', 'review_answer_timestamp']
for feature in datetime_features_reviews:
    reviews_i[feature] = pd.to_datetime(reviews_i[feature], errors='coerce')
reviews_i.dtypes
```

```python
orders_i["estimated_delivery_time"] = (orders_i["order_estimated_delivery_date"] - orders_i["order_purchase_timestamp"]).dt.days

actual_avg = orders_i["delivery_time"].mean()
estimated_avg = orders_i["estimated_delivery_time"].mean()

print(f"Average Actual Delivery Time: {actual_avg:.2f} days")
print(f"Average Estimated Delivery Time: {estimated_avg:.2f} days")
```

#### % DELIVERED EARLY vs. ON TIME vs. LATE

```python
orders_i["delivery_performance"] = np.where(
    orders_i["order_delivered_customer_date"] < orders_i["order_estimated_delivery_date"], "Early",
    np.where(
        orders_i["order_delivered_customer_date"] == orders_i["order_estimated_delivery_date"], "On Time",
        "Late"
    )
)

performance_dist = orders_i["delivery_performance"].value_counts(normalize=True) * 100
plt.pie(performance_dist, labels=performance_dist.index, colors=['mediumturquoise', 'yellowgreen'], autopct='%1.1f%%')
plt.title("Delivery performance")
```

#### MEDIAN vs MEAN DELIVERY DELAY

```python
mean_delay = orders_i["delivery_delay"].mean()
median_delay = orders_i["delivery_delay"].median()

print(f"Mean Delivery Delay: {mean_delay:.2f} days")
print(f"Median Delivery Delay: {median_delay:.2f} days")
```

<div class='alert alert-block alert-success'>
    This shows that the parcels are almost 12 days early.
</div>

### PAYMENTS

#### PAYMENT TYPE DISTRIBUTION

```python
payment_counts = payments_i['payment_type'].value_counts()

plt.figure(figsize=(6,6))
plt.bar(payment_counts.index, payment_counts.values, color='limegreen')
plt.title("Payment Type Distribution")
plt.xlabel("Payment Type")
plt.ylabel("Number of transactions")
plt.xticks(rotation=45)
plt.show()
```

#### INSTALLMENTS vs PAYMENT VALUE

```python
installment_summary = payments_i.groupby('payment_installments')['payment_value'].mean()

plt.figure(figsize=(9,6))
plt.bar(installment_summary.index, installment_summary.values, color='darksalmon')
plt.xlabel("Number of Installments")
plt.ylabel("Payment Value")
plt.title("Installments vs. Payment Value")
plt.show()
```

### REVIEWS

#### REVIEW SCORES DISTRIBUTION

```python
agg = reviews_i['review_score'].value_counts().sort_index()
agg
```

```python
reviews_i['sentiment'] = reviews['review_score'].apply(lambda x : 'Positive' if x>=4 else ('Neutral' if x==3 else 'Negative'))
sentiment_dist = reviews_i['sentiment'].value_counts(normalize=True)*100
sentiment_dist
```

```python
plt.figure(figsize=(8,6))
plt.pie(sentiment_dist, labels = sentiment_dist.index, autopct='%1.1f%%', colors=['mediumpurple','cornflowerblue', 'teal'])
plt.title("Reviews Scores Distribution")
plt.show()
```

### CROSS-DATASET INSIGHTS

#### REVIEWS TIMING DISTRIBUTION

```python
reviews_timing = reviews_i.merge(orders[['order_id', 'order_purchase_timestamp']],
                                 on = 'order_id',
                                 how = 'left')
```

```python
reviews_timing["review_creation_date"] = pd.to_datetime(reviews_timing["review_creation_date"])
reviews_timing["order_purchase_timestamp"] = pd.to_datetime(reviews_timing["order_purchase_timestamp"])
```

```python
reviews_timing["review_delay"] = (reviews_timing["review_creation_date"] - reviews_timing["order_purchase_timestamp"]).dt.days
reviews_timing["review_delay"].describe()
```

```python
plt.figure(figsize=(8,6))
plt.hist(reviews_timing["review_delay"], bins=30, color='gold')
plt.xlabel("No. of Days after Purchase")
plt.ylabel("Number of Reviews")
plt.title("Distribution Review Delays")
plt.show()
```

#### INSTALLMENTS vs REVIEWS

```python
data = payments_i.merge(reviews_i[["order_id", "review_score"]],
                        on = "order_id",
                        how = "inner")
```

```python
ins_vs_rev = data.groupby("payment_installments")["review_score"].mean().reset_index()

plt.figure(figsize = (8,6))
plt.plot(ins_vs_rev["payment_installments"], ins_vs_rev["review_score"], marker="o", color='darkorchid')
plt.xlabel("Number of Installments")
plt.ylabel("Average Review Score")
plt.title("Installments vs. Average Review Score")
plt.show()
```
